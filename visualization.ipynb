```python
# Import the necessary libraries

%matplotlib inline
import matplotlib as mpl
mpl.style.use('ggplot')
import matplotlib.pyplot as plt

import numpy as np
```

```python
# Creating a simple graph with matplotlib

x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10

fig, ax = plt.subplots() # Generating the Figure and Axes instances
ax.plot(x, y1, color='blue', label='y(x)') # Creating the actual graphs
ax.plot(x, y2, color='red', label="y'(x)")
ax.plot(x, y3, color='green', label='y"(x)') 
ax.set_xlabel('x') # set the label for the x-axis
ax.set_ylabel('y') # set the label for the y-axis
ax.legend() # adding the legend
```
